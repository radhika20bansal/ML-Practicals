```python
import numpy as np 
import pandas as pd
```

```python
df = pd.read_csv("C:\\Users\Radhika Bansal\Downloads\Train.csv")
```

```python
df.head(10)
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df['Item_Weight']= df['Item_Weight'].fillna(np.mean(df['Item_Weight']))
```

```python
print(df['Item_Fat_Content'].value_counts())
print(df['Item_Type'].value_counts())
print(df['Outlet_Identifier'].value_counts())
print(df['Outlet_Size'].value_counts())
print(df['Outlet_Location_Type'].value_counts())
print(df['Outlet_Type'].value_counts())
```

```python
df = pd.get_dummies(df,columns=['Item_Fat_Content','Item_Type','Outlet_Identifier','Outlet_Size','Outlet_Location_Type','Outlet_Type'])
```

```python
df
```

```python
df['Item_Identifier'] = df['Item_Identifier'].astype('category')
df.dtypes
```

```python
df['Item_Identifier'] = df['Item_Identifier'].cat.codes
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaled_df = scaler.fit_transform(df)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(scaled_df)
```

```python
x_pca = pca.transform(scaled_df)
```

```python
x_pca.shape
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0],x_pca[:,1], cmap='plasma')
plt.xlabel("First Principal Component")
plt.ylabel("Second Principal Component")
```

```python
df_comp = pd.DataFrame(pca.components_, columns=df.columns)
```

```python
import seaborn as sns
plt.figure(figsize=(12,6))
sns.heatmap(df_comp,cmap='plasma')
```
